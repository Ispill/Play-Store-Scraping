## Librerías

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
from scipy import stats
```

## Leer CSV

```python
# Cargar el dataset (supongo que ya lo tienes cargado como df, pero lo incluyo por si acaso)
df = pd.read_csv("./source/googleplaystore.csv")
```

# Limpiar los datos (eliminación de nulos y transformación de columnas)

```python
df = df.dropna()
df['Installs'] = df['Installs'].str.replace('+', '').str.replace(',', '').astype(int)
df['Reviews'] = pd.to_numeric(df['Reviews'], errors='coerce')
```

## Análisis ANOVA

```python
# Agrupamos las descargas por categoría
categories = df['Category'].unique()
grouped_data = [df[df['Category'] == category]['Installs'] for category in categories]

# Realizamos el ANOVA de un solo factor para comparar las descargas entre categorías
anova_result = stats.f_oneway(*grouped_data)

# Mostramos los resultados del ANOVA
print("Resultado ANOVA:")
print(f"F-value: {anova_result.statistic}")
print(f"p-value: {anova_result.pvalue}")
```

## Correlación de Pearson

```python
# Calculamos la correlación de Pearson entre descargas e reseñas en el dataset completo
pearson_corr, pearson_pvalue = stats.pearsonr(df['Installs'], df['Reviews'])

# Mostramos los resultados de la correlación de Pearson
print("\nCorrelación de Pearson:")
print(f"Coeficiente de correlación (r): {pearson_corr}")
print(f"p-value: {pearson_pvalue}")
```

```python
category = df["Category"].value_counts().sort_index().sort_values()

# Crear el gráfico de barras
fig = px.bar(x=category.index, y=category.values, labels={'x':'Generos', 'y':'Cantidad de aplicaciones'}, width=1200, height=800)

# Agregar los valores encima de cada barra
fig.update_traces(text=category.values, textposition='outside')

# Etiquetas y título
fig.update_layout(title='Cantidad de aplicaciones por genero', title_x=0.5)
```

```python
# Visualización de la cantidad de descargas por categoría
category_installs = df.groupby('Category')['Installs'].mean().sort_values()

fig = px.bar(category_installs, labels={'x':'Categoría', 'y':'Descargas Promedio'},
             title='Descargas promedio por categoría', width=1200, height=600)
fig.show()

# Visualización de la relación entre descargas y reseñas (scatter plot)
fig2 = px.scatter(df, x='Installs', y='Reviews', color='Category',
                  title='Relación entre descargas y reseñas por categoría',
                  labels={'Installs': 'Número de Instalaciones', 'Reviews': 'Número de Reseñas'},
                  log_x=True, log_y=True, width=1200, height=800)
fig2.show()
```
